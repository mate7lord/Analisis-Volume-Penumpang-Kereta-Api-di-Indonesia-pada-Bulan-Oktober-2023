```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
file_path = 'Volume Penumpang Kereta Api per Oktober 2023 PT Kereta Api Indonesia.xlsx'
data = pd.read_excel(file_path, sheet_name='Vol Stasiun')
```

```python
# Tampilkan beberapa baris pertama data
print(data.head())
```

```python
# Bersihkan dan persiapkan data
# Menghapus baris-baris yang tidak diperlukan
data = data.drop([0, 1, 2])  # Menghapus tiga baris pertama yang tidak diperlukan
data.columns = data.iloc[0]  # Mengambil baris ke-4 sebagai header
data = data.drop(3)  # Menghapus baris ke-4 setelah dijadikan header
data = data.reset_index(drop=True)  # Mereset index
```

```python
# Ganti nama kolom untuk kemudahan akses
data.columns = ['No', 'Stasiun', 'Singkatan', 'Wilayah', 'Transaksi',
                '1_Okt', '2_Okt', '3_Okt', '4_Okt', '5_Okt', '6_Okt',
                '7_Okt', '8_Okt', '9_Okt', '10_Okt', '11_Okt', '12_Okt',
                '13_Okt', '14_Okt', '15_Okt', '16_Okt', '17_Okt', '18_Okt',
                '19_Okt', '20_Okt', '21_Okt', '22_Okt', '23_Okt', '24_Okt',
                '25_Okt', '26_Okt', '27_Okt', '28_Okt', '29_Okt', '30_Okt',
                '31_Okt', 'Total_Volume']
```

```python
# Ubah tipe data kolom volume penumpang ke numerik
for col in data.columns[5:-1]:
    data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
# Analisis dan Visualisasi
# Contoh visualisasi tren volume penumpang harian untuk seluruh stasiun
daily_volume = data.iloc[:, 5:-1].sum(axis=0)
daily_volume.plot(kind='line', figsize=(15, 7))
plt.title('Tren Volume Penumpang Harian di Bulan Oktober 2023')
plt.xlabel('Tanggal')
plt.ylabel('Volume Penumpang')
plt.grid(True)
plt.show()
```
